```python
di = 27 * BIN_FREQ_PER_DAY
print(di)
di = round(di)
print(di)
```

```python
errors = []
for i in range(di, len(df.index)):
    assert i - di >= 0
    error = df.iloc[i].Vp_obs - df.iloc[i - di].Vp_obs
    errors.append(error)

errors = np.array(errors)
errors = errors[np.isfinite(errors)]
```

```python
'%.2f'%np.sqrt(np.mean(np.square(errors)))
```
